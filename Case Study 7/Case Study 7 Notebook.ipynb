## Data Cleaning

### Missing Values

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))


columns_with_nan = df.columns[df.isna().any()].tolist()

if len(columns_with_nan) > 0:
    # Calculate the proportion of missing values for each column
    missing_proportions = df[columns_with_nan].isna().sum() / len(df)

    # Plot the proportions using a horizontal bar plot
    ax = missing_proportions.plot(kind="barh", title="Proportion of Missing Values by Features")
    ax.set_xlabel('Proportion')  # Set the x-axis label
    ax.set_ylabel('Features')
    plt.show()
else:
    print("There are no columns with Missing Values.")
```
